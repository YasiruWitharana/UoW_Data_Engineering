```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
filtered_petal_length = iris[iris['petal_length']>1.5]
display(filtered_petal_length)
```

```python
filtered_petal_width = iris[iris['petal_width']<0.4]
display(filtered_petal_width)
```

```python
from operator import irshift
filtered_setosa = iris[(iris['species']=='setosa') & (iris['sepal_width'] > 3.0)] # if we are using multiple condition we use brackets() and &
print('\n Setosa flower with sepal width greater than 3cm')
display(filtered_setosa)
```

```python
filtered_species = iris[iris['species'].isin(['versicolor','virginica'])]
print('\n Flowers that are either Verscolor or Virginica:') # we can multiple values in one column
display(filtered_species)
```

```python
filtered_petal_width = iris.query('petal_width < 0.5')
print('\n Flowers with petal width less than 0.5cm')
display(filtered_petal_width)
```

```python
filtered_species_name = iris[iris['species'].str.contains('a')]
print('\n Flowers whose species name contains "a": ')
display(filtered_species_name)
```

```python
random_sample = iris.sample(frac = 0.2, random_state=42)
print('\n Random Sample (20% of the data):')
display(random_sample)
```
